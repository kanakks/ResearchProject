```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Australia_.csv')
```

```python
df.head()

```

```python
df.info()
# df.describe()
# df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x='Year', y='GDP(in US $)', label='GDP')
plt.title('Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x='Year', y='Energy Consumption(in TJ)', label='Energy Consumption')
plt.title('Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
# sns.lineplot(data=df, x='Year', y='Energy Consumption(in TJ)', label='Energy Consumption')
sns.lineplot(data=df, x='Year', y='CO2 Emission(in kt)', label='CO2 Emissions')
plt.title('Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
